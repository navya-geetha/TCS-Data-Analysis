# **TCS Stock Market Analysis**

**Importing Required Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

**Loading the Dataset**

```python
# Load CSV file
df = pd.read_csv("/content/TCS_stock_action.csv")

# Display first 5 rows
print(df.head())
```

```python
# Basic Information
print(df.info())
print(df.describe())
```

**Data Cleaning**

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

```python
df = df.drop_duplicates()
```

**Exploratory Data Analysis (EDA)**

```python
print(df.corr())
```

**Data Visualization**

```python
plt.figure()
plt.plot(df['Date'], df['Dividends'])
plt.title("TCS Dividends Over Time")
plt.xlabel("Date")
plt.ylabel("Dividends")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
